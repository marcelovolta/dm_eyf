```python
# Instalar sympy si no está presente
!pip install sympy --quiet

import random
from sympy import primerange

# Rango de números primos
start = 100_003
end = 999_983

# Generar todos los primos en el rango
primos = list(primerange(start, end + 1))

# Elegir 5 al azar sin repetición
random.seed(60)
seleccionados = random.sample(primos, 5)

# Mostrar resultados
print("5 números primos seleccionados al azar:")
print(seleccionados)
```
